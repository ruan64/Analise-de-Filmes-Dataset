```r
#Criação da função men_ci
mean_ci <- function(x, conf = 0.95) {
  m <- mean(x, na.rm = TRUE)
  se <- sd(x, na.rm = TRUE) / sqrt(sum(!is.na(x)))

  t_crit <- qt((1 + conf) / 2, df = sum(!is.na(x)) - 1)
  ymin <- m - t_crit * se
  ymax <- m + t_crit * se
  data.frame(y = m, ymin = ymin, ymax = ymax)
}
```

```r
#Criar Gráfico de dispersão
gs <-ggplot(data = dados) +
#Gerando as coordenadas dos pontos
  geom_point(aes(y = LucroMundial , x = LucroLocal , color = "red"), shape = 18, size = 2) +
#Gerando a linha do Grafico
  geom_smooth(aes(y = LucroMundial , x = LucroLocal ),method = "lm", se = F, color = "black")+
  labs(y = "Lucro Mundial", x = "Lucro Local", title = "Lucro Local por Lucro Mundial")+
#Formatando o Gráfico
#Tirando os numeros de notação científica
#colocando manualmente do sibolo de dólar
#Indicando que os números é em Milhões
  scale_y_continuous(labels = scales::number_format(big.mark = ".", decimal.mark = ",",
                                                    prefix = " US$ ",
                                                    scale = 1e-6, suffix = " M"))+
  scale_x_continuous(labels = scales::number_format(big.mark = ".",decimal.mark = ",",
                                                    prefix = " US$ ",
                                                    scale = 1e-6, suffix = " M"))+
#Mudando o Tema do grafico
                                                    theme_dark()
  gs


```

```r
#Gerando Gráfico de colunas
ggplot(data = dados) +
#Fazendo o historico de orçamento
  geom_histogram(aes(x = Orcamento), bins = 30, color = "black") +
#Nomes para o eixo X e Y
  labs(y = "Quantidade de Filmes", x = "Orçamento", title = "Quantidade de Filmes por Orçamento", caption = "Fonte : FiveThirtyEight")+
#Convertendo números de notação cientifica para números em Milhões
  scale_x_continuous(labels = scales::number_format(big.mark = ".",decimal.mark = ",",
                                                    prefix = " US$ ",
                                                    scale = 1e-6, suffix = " M"))+
#Barras sem espaçamento na parte inferior e com 100 de espaçamento na superior
  scale_y_continuous(expand = expansion(add = c(0,100)))+
#Mudando o tema do Gráfico
  theme_dark()
```

```r
#Gerando o Grafico de barras
ggplot(data = dados) +
  geom_bar(aes(x = Genero, y = ..count..), color = "black")+
#Ajustando novamento o espaçamento
  scale_y_continuous(expand = expansion(add = c(0,100)))+
#Corrigindo o nome dos gêneros dos filmes
  scale_x_discrete(labels = c("Ação","Aventura","Comédia","Drama","Terror"))+
#Nomeando o eixos
  labs(y = "Quantidade de Filmes", x = "Gênero",
#Adicionando o Titulo
       title = "Quantidade de Filmes por Gênero",
#Adicionando a Fonte
       caption = "Fonte : FiveThirtyEight")+
#Mudando o Tema do gráfico
       theme_dark()
```
